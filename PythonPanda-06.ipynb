```python
import pandas as pd
```

```python
df1=pd.DataFrame({
    "city":["new york",'chicago','orlando'],
    "temperature":[21,42,24]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humidy":[64,76,59]
})
df2
```

```python
 #To merge dataframes use the following command (Merge happens for all rows when total number of columns and value are equal (intersection operation))
df=pd.merge(df1,df2,on='city')
```

```python
df
```

```python
df3=pd.DataFrame({
    "city":["Bangalore","Delhi","Mumbai","Hyderabad","Chennai"],
    "temperature":[23,45,24,32,42]
})
df3
```

```python
df4=pd.DataFrame({
    "city":["Bangalore","Delhi","Mumbai","Hyderabad"],
    "Humidity":[23,25,28,29]
})
df4
```

```python
# When we merge two different dataframes with different no.of rows present it will perform intersection operation.
df5=pd.merge(df3,df4,on='city')
df5
```

```python
# To perform merge with all values use the following argument (how='outer': Union operation), when any of the value is missing in any of rows it will place NAN
df5=pd.merge(df3,df4,on='city',how='outer')
df5
#(OR)
df5=pd.merge(df3,df4,on='city',how='left')
df5
```

```python
df5=pd.merge(df3,df4,on='city',how='right')
df5
```

```python
#To know from which data frame the data has come whether both or left only or right only use following argument in the merge method
df5=pd.merge(df3,df4,on='city',how='outer',indicator=True)
df5
```

```python
df6=pd.DataFrame({
    "city":["Bangalore","Chennai","Mumbai","Delhi"],
    "temperature":[32,27,38,29],
    "humidity":[23,28,29,31]
})
df6
```

```python
df7=pd.DataFrame({
    "city":["Delhi","Chennai","Mumbai","Ranchi"],
    "temperature":[45,23,28,53],
    "humidity":[23,53,24,28]
})
df7
```

```python
#When we merge two dataframes with similar column names dataframe differentiate the results as following
pd.merge(df6,df7,on='city')
```

```python
#When we merge two dataframes with similar column names dataframe differentiate the results as following, to avoid default suffixes use the argument in the method
pd.merge(df6,df7,on='city',suffixes=('_left','_right'))
```
